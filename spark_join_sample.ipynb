```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
# read all data frames

df_education = spark.read \
               .option("header", "true") \
               .option("inferSchema", "true") \
               .csv("Files/Education_Details.csv")
df_education.show()

df_salary = spark.read \
            .option("header", "true") \
            .option("inferSchema", "true") \
            .csv("Files/Salary_Details.csv")
df_salary.show()
```

```python
df_education.join(df_salary,on="Education Level", how="inner").show()
```

```python
df_education.join(df_salary,on="Education Level", how="left").show()
```

```python
df_education.join(df_salary,on="Education Level",how="right").show()
```

```python
df_education.join(df_salary,on="Education Level",how="outer").show()
```

```python
df_education.join(df_salary,on="Education Level", how="leftsemi").show()
```

```python
df_education.join(df_salary,on="Education Level",how="leftanti").show()
```

```python
df_education.join(df_salary,on="Education Level",how="cross").show()
```
