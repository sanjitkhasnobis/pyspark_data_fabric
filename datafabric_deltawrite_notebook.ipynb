# Preapre Data

```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql import Row

data = [Row(id=1, name="Alice"), Row(id=2, name="Bob")]

new_data = [
    Row(id=2,name="Bobby"),
    Row(id=3,name="Charlie")
]

df_employees = spark.createDataFrame(data)

df_employees_new_data = spark.createDataFrame(new_data)
```

# Create Delta Table

```python
df_employees.write.format("delta").mode("overwrite").saveAsTable("df_delta_employees_tab")
```

# Insert/Update New Data

```python
from delta.tables import DeltaTable

delta_table = DeltaTable.forPath(spark, "abfss://ac7781b8-403d-4f5e-8354-17df61045427@onelake.dfs.fabric.microsoft.com/9943f12b-9887-4c79-b67a-07a542560d2a/Tables/df_delta_employees_tab")

delta_table.alias("target").merge(
    df_employees_new_data.alias("source"),
    "target.id = source.id"
).whenMatchedUpdate(set={"name": "source.name"}
).whenNotMatchedInsert(values={
    "id": "source.id",
    "name": "source.name"
}).execute()
```
